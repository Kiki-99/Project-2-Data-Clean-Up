```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
# Name of the CSV File
ins_file_2010 = "Resources/2010_US_Ins_Cov.csv"
ins_file_2012 = "Resources/2012_US_Ins_Cov.csv"
ins_file_2014 = "Resources/2014_Us_Ins_Cov.csv"
ins_file_2016 = "Resources/2016_Us_Ins_Cov.csv"
ins_file_2018 = "Resources/2018_Us_Ins_Cov.csv"
```

```python
# Read CSV Files
ins_df_2010 = pd.read_csv(ins_file_2010, names=["Location", "Employer", "Non-Group", "Medicaid", "Medicare",
       "Military", "Uninsured", "Total"], header=0)
ins_df_2012 = pd.read_csv(ins_file_2012, names=['Location', 'Employer', 'Non-Group', 'Medicaid', 'Medicare',
       'Military', 'Uninsured', 'Total'], header=0)
ins_df_2014 = pd.read_csv(ins_file_2014, names=['Location', 'Employer', 'Non-Group', 'Medicaid', 'Medicare',
       'Military', 'Uninsured', 'Total'], header=0)
ins_df_2016 = pd.read_csv(ins_file_2016, names=['Location', 'Employer', 'Non-Group', 'Medicaid', 'Medicare',
       'Military', 'Uninsured', 'Total'], header=0)
ins_df_2018 = pd.read_csv(ins_file_2018, names=['Location', 'Employer', 'Non-Group', 'Medicaid', 'Medicare',
       'Military', 'Uninsured', 'Total'], header=0)
```

## US Insurance Coverage 2010 : Health Insurance Coverage of Females 19-64

```python
ins_df_2010
```

```python
# check datatypes
ins_df_2010.dtypes
```

```python
#convert values to percentages and then use .style.format({"column name" : "{:,.2f}%"})
employer_per = ins_df_2010["Employer"] * 100
employer_per

non_group_per = ins_df_2010["Non-Group"] * 100
non_group_per

medicaid_per = ins_df_2010["Medicaid"] * 100
medicaid_per

medicare_per = ins_df_2010["Medicare"] * 100
medicare_per

military_per = ins_df_2010["Military"] * 100
military_per

uninsured_per = ins_df_2010["Uninsured"] * 100
uninsured_per

total_per = ins_df_2010["Total"] * 100
total_per
```

```python
# assign locations to list for new dataframe
locations = ins_df_2010["Location"]
locations
```

```python
# make a new dataframe with percentages and without the footnotes column
us_ins_2010_df = pd.DataFrame({'Location': locations,
                               'Year': 2010,
                               'Employer': employer_per, 
                               'Non-Group': non_group_per,
                               'Medicaid': medicaid_per,
                               'Medicare': medicare_per, 
                               'Military': military_per,
                               'Uninsured': uninsured_per, 
                               'Total': total_per})
us_ins_2010_df
```
